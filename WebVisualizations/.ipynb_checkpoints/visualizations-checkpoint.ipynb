```python
#imports
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
cities_df=pd.read_csv('Resources/cities.csv')
cities_df.head()
```

```python
#temperature vs latitude:
plt.scatter(cities_df.Lat,cities_df["Max Temp"])
plt.title("City Latitude vs. Max Yearly Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
plt.savefig('Tempvslatitude')
```

```python
#humidity vs latitude
plt.title("Humidity vs City Latitude")
plt.scatter(cities_df.Lat,cities_df.Humidity)
plt.xlabel("Latitude")
plt.ylabel("Humidity %")
plt.savefig('Humidvslat')
```

```python
#wind speed vs latitude
plt.scatter(cities_df.Lat,cities_df["Wind Speed"])
plt.title("Wind Speed vs City Latitude")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.savefig('Windvslat')
```
